```python
from cgi import test
import math
import numpy as np
import copy
import random as rd
from scipy import special
import network


N=1e2
input_len=4
output_len=4
test_data=[]
test_label=[]

for i in range(int(N)):
    test_data.append(np.array([0.0]*input_len))
    min=300
    min_index=0
    for j in range(len(test_data[i])):
        test_data[i][j]=rd.randint(-10,10)
        if test_data[i][j]<min:
            min=test_data[i][j]
            min_index=j
    test_label.append(np.array([0.0]*output_len))
    test_label[i][min_index]=1.0

net=network.network([input_len,10,output_len])
net.alpha=1e-2

net.train(test_data,test_label,500)

net.predict(np.array([10,20,30,40]))
net.printResult()
```
